```python

def limit_voronoi_with_convex_hull(vor , hull):
    hull_polygon = Polygon(points[hull.vertices])

    # For each infinite Voronoi ridge
    for simplex in vor.ridge_vertices:
        if -1 in simplex:  # This is an infinite ridge
            # Find the two points of the ridge
            p1, p2 = [vor.vertices[i] for i in simplex if i != -1]

            # Find the direction of the infinite ridge
            t = vor.points[simplex[1]] - vor.points[simplex[0]]
            t /= np.linalg.norm(t)

            # Create a line that goes through p1 and has the direction t
            # We multiply t by a large number to ensure it goes outside the convex hull
            line = LineString([p1, p1 + 1e5 * t])

            # Find the intersection of the line with the convex hull
            intersection = hull_polygon.intersection(line)

            if intersection:
                # Replace the infinite point with the intersection point
                vor.vertices[simplex[simplex.index(-1)]] = np.array(intersection.coords[1])

    return vor

def assign_to_clusters(dataset, vor):
    """
    Assigns each data point in dataset to a cluster based on the Voronoi region it belongs to.

    Parameters:
    - dataset: list of data points
    - vor: Voronoi object

    Returns:
    - Clusters: List of clusters, where each cluster is a list of data points
    """
    Clusters = [[] for _ in vor.regions]

    for pt in dataset:
        # Find the Voronoi region the point belongs to
        min_distance = float('inf')
        closest_region = -1
        for i, ridge in enumerate(vor.ridge_points):
            distance = np.linalg.norm(np.array(pt) - vor.points[ridge[0]])
            if distance < min_distance:
                min_distance = distance
                closest_region = vor.point_region[ridge[0]]
        if closest_region != -1:
            Clusters[closest_region].append(pt)
    print(Clusters)
    return [cluster for cluster in Clusters if cluster]  # filter out empty clusters

def plot_clusters(Clusters):
    """
    Plots the clusters in different colors.

    Parameters:
    - Clusters: List of clusters, where each cluster is a list of data points
    """
    plt.figure()
    colors = plt.cm.jet(np.linspace(0, 1, len(Clusters)))
    for i, cluster in enumerate(Clusters):
        x_vals, y_vals = zip(*cluster)
        plt.scatter(x_vals, y_vals, color=colors[i], label=f'Cluster {i + 1}')

    plt.title("Clusters based on Voronoi regions")
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()



# Helper function to check if a point is inside CH
def is_inside_CH(point, tri):
    return tri.find_simplex(point) >= 0

```
